```python
#######################
# _____ Soldier class_____
######################

class Soldier:
    
    def __init__(self,name,land):
        self.__level = 1
        self.__name = name
        self.__land = land
        self.__blood = 100
        self.__weapon = None # should be a list 
        self.__pionts = 0 # every 5 pionts level up
        self.__inventory = {'food':5,'medicine':3,'armor':0,'weapon':0}
        self.__str_of_str = 0 #strength of strike 
        print("Soldier", self.__name, "You have a heay burden you shall carry...")
        print("Your people have supplied you with 5 amounts of food and 3 doses of medice. Use them wisely")
        print("Make a weapon")
        print("*************************************************")
        
    def calc_srt(self,land):
        if(self.__land == land):
            return self.__weapon.getPower() * 0.5 * self.__level
        else:
            return self.__weapon.getPower() * 0.25 * self.__level
        
    def inc_pionts(self,amount):
        self.__pionts = self.__pionts + amount
        if (self.__pionts >= 5):
            self.__pionts = 0
            self.__level += 1
            print(self.__name,"You are one step higher Soldier!")
        print(self.__name,"your new pionts now are:",self.__pionts)   
            
    def lose_fight(self):
        self.__blood -= 50
        if(self.__blood<=0):
            print("Rest in peace soldier",self.__name)
            # need to desroy the instance
        else:
            print("soldier", self.__name,"your enemy has defeated you...Be carefull your blood rate is", self.__blood)
    def assign_waepon(self, weapon):
        self.__weapon = weapon
        self.__inventory['weapon'] = self.__inventory['weapon'] + 1
            
    def recovery(self):
        if(self.__inventory['medicine'] > 0 and self.__blood < 100):
            self.__inventory['medicine'] = self.__inventory['medicine'] - 1
            self.__blood = self.__blood + 50
            print("blood restored")
            print("Blood = ", self.__blood)
            print("Your medicine storage is ",self.__inventory['medicine'])
        elif(self.__blood == 100):
            print("soldier", self.__name,"You are very healthy soldier, move on!")
        elif():
            print("soldier", self.__name,"Go get some midicine sypply!")
            
    def getName(self):
        return self.__name
            
#######################
# _____ Weapon class_____
######################
    
class Weapon:
    
    def __init__(self, t, power):
        self.__type = t
        self.__power = power
    
    def getPower(self):
        return self.__power
        
    
    
    
    
        
def fight(s1,s2,land):
    s1_str_of_str = s1.calc_srt(land)
    s2_str_of_str = s2.calc_srt(land)
    if(s1_str_of_str > s2_str_of_str):
        s1.inc_pionts(3)
        print(s1.getName(),"have won!")
        s2.lose_fight()
    elif(s1_str_of_str < s2_str_of_str):
        s2.inc_pionts(3)
        print(s2.getName(),"have won!")
        s1.lose_fight()
    else:
        s2.lose_fight()
        s1.lose_fight()
        print("No winners on this fight")
    print("*************************************")

        
       
        
        
        
```

```python
#######################
# _____ main_____
######################

def main():
    
    s1 = Soldier("Maha","earth")
    s2 = Soldier("Fatimah","water")
    
    w1 = Weapon("sord",10)
    w2 = Weapon('knife',5)
    
    s1.assign_waepon(w1)
    s2.assign_waepon(w2)
    
    fight(s1,s2,'Earth')
    
    s2.recovery()

    
    
main()





```
